**Análisis exploratorio de datos EA FC25**

Resumen:
La base de datos tiene un formato csv, en donde se encuentras estadisticas de jugadores del popular videojuego EA 25 (antiguamente FIGA

1. Importar librerías

```python
import numpy as np #Manejo de arreglos (Alg. líneal en python)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Importar datos

La primera forma es importarlos localmente

```python
datos = pd.read_csv('/content/male_players.csv')
datos.head(10)
```

```python
datos.info()
```

```python
datos.describe()
```

```python
datos.shape
```

Esto significa que contamos con 16161 muestras o samples y 57 columnas o variables

```python
#Eliminar la columna indice
datos.drop('Unnamed: 0', axis=1, inplace = True)
```

**10 Ligas con mayor cantidad de jugadores**

```python
ligas = datos["League"].value_counts().head(10)
```

```python
#Gráfica
plt.figure(figsize=(6, 4))
ligas.plot(kind="bar", color = "salmon")
plt.title("Ligas con mayor cantidad de jugadores")
plt.xlabel("ligas")
plt.ylabel("cantidad de jugadores")
plt.show()
```

Distribución del "Overall

```python
x = datos["OVR"]
bins_ = np.arange(40,100,2)
bins_
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(x = "OVR",
             data = datos,
             bins = bins_,
             kde = True,
            color = 'salmon')

plt.title("Distribución del Overall")
plt.grid(True)
plt.show()
```

Overall por posición

```python
posiciones = datos.groupby('Position')['OVR'].median().sort_values().index
posiciones
```

##Diagrama de cajas (Boxplot)


```python
sns.boxplot(x = 'Position',
            y = 'OVR',
            data = datos,
            order = posiciones,
            palette = 'Accent')
plt.title('Overall por posicion')
plt.show()
```
